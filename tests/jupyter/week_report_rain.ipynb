最近一周降水空间分布对比

```python
dir_ob = "SURFACE/RAIN24_NATIONAL_08-08/YYYYMMDDHH0000.000"
dir_r = "NWFD_SCMOC/RAIN24/YYMMDDHH.TTT"
now = datetime.datetime.now()
today = datetime.datetime(now.year,now.month,now.day,8,0)
station = meb.read_station(meb.station_国家站)
```

```python
ip,port = meb.io.read_gds_ip_port(para["ip_port_file"])
meb.print_gds_file_values_names(ip,port,"SURFACE/RAIN24_NATIONAL_08-08/20200226080000.000")
```

```python
veri_day_count = 7
time1 = today - datetime.timedelta(days = veri_day_count)
ob_list = []
for i in range(veri_day_count):
    time2 = time1 + datetime.timedelta(days = i)
    path = meb.tool.path_tools.get_path(dir_ob,time2)
    sta = mep.application.read_stadata(para,path,meb.gds_element_id.降水_24小时,station)
    ob_list.append(sta)
```

```python
#读取网格预报
dir_gridfore = r"NWFD_SCMOC/RAIN24/YYMMDDHH.TTT"
gridfore_list_list=[]
gf_sta_list = []
for i in range(veri_day_count):
    gf_list = []
    for d in range(3):
        dh = d * 24 + 24
        time2 = time1 + datetime.timedelta(days = i)
        time3 = time2 - datetime.timedelta(hours = dh)
        path = meb.path_tools.get_path(dir_gridfore,time3,dh)
        grd = mep.application.read_griddata(para,path)
        meb.set_griddata_coords(grd,member_list = ["NWFD_SCMOC"])
        sta = meb.fun.interp_gs_nearest(grd,station)
        gf_list.append(grd)
        gf_sta_list = [sta]
    gridfore_list_list.append(gf_list)
```

```python
dir_ec = r"ECMWF_HR/APCP/YYMMDDHH.TTT"
ec_list = []
```

```python
ob_all = pd.concat(ob_list,axis = 0)
ob_all = meb.between_value_range(ob_all,0,1000)
```

```python
dat = ob_all.loc[:,"data0"].values
dat_rain = dat[dat>0]
dat_rain.sort()
index = int(dat_rain.size/20)
grade =  round(dat_rain[-index], 1)
print("最近一周强度前5%的降水大于"+str(grade)+"mm")
```

```python
fre_grade = mep.score(ob_all,mem.hap_count,group_by = "time",para1 = [grade])
xticks = mep.base.fun.get_time_str_list(fre_grade[1])
```

```python
fig = plt.figure(figsize = (10,4))
x = np.arange(veri_day_count)
plt.plot(x,fre_grade[0])
plt.xticks(x,xticks)
plt.xlabel("时间")
plt.ylabel("相对较强降雨的站次")
```

```python
max_index = np.argmax(fre_grade[0])
sta = ob_list[max_index]
grd_list = gridfore_list_list[max_index]
for i in range(3):
    grd = grd_list[i]
    mep.base.rain_24h_comprehensive_chinaland_sg(sta,grd)
```
